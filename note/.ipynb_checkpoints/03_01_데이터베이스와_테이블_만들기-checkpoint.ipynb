## 데이터베이스 생성하기

```bash
$ mysql-ctl cli

mysql> show databases;

CREATE DATABASE <name>;

e.g., 
- CREATE DATABASE soap_store; # snake case 추천
- CREATE DATABASE DogApp;

```

SQL 명령어에 대문자 사용 이유? >> 나중에 설명
`CREATE DATABASE testing_db`
```
오라클, MySQL, MSSQL등 온갖 DB에서 쿼리문을 작성할 경우
대부분은 모든 쿼리를 대문자로 작성하는 경향이 있는데
프로그래밍 언어는
대문자 소문자 합치기, 스네이크 케이스, 파스칼 케이스 등
온갖 방법이 있지만 SQL은 그런거 없고
그냥 다 대문자를 쓰게 된다


근데 왜 굳이 대문자로만 쓰냐면
옛 쿼리 에디터는 지금처럼 구문 강조나 색깔 변경 등
가독성을 높여주기 위한 장치가 전혀 없었기 때문에
가독성을 위한 최적의 방법이
모든 글자를 대문자로 써버리는 거였다

 
이후 시간이 꽤 흘러
현재는 대부분의 SQL 에디터가 강조 표시, 색깔 변경을 지원하지만
쿼리는 옛날부터 대문자로 써왔고 그게 쭉 이어져 내려와서
지금까지도 모든 쿼리는 대문자로 작성하게 되는 것이다.
```



## 테이블 소개

## 기본 Data types

cats 테이블에서 세 개의 열에 3개의 헤더가 있음(이름, 품종, 나이)
데이터가 일관적이지 않다. 나이의 테이블 자료형에 문자가 있는 것은 이상하다.
SQL은 이를 강제한다.

테이블을 생성할 때 어떤 자료형을 찾고 있는지 명시해야 한다.


Numeric Types
- INT : 정수 표현
- SMALLINT
- TINYINT
- MEDIUMINT
- BIGINT
- DECIMAL
- NUMERIC
- FLOAT
- DOUBLE
- BIT

String Types
- CHAR
- VARCHAR : 텍스트 표현
- BINARY
- VARBINARY
- BLOB
- TINYBLOB
- MEDIUMBLOB
- LONGBLOB
- TEXT
- TINYTEXT
- MEDIUMTEXT
- LOBGTEXT
- ENUM

Data Types
- DATE
- DATETIME
- TIMESTAMP
- TIME
- YEAR

대부분의 경우 모든 것을 사용하지는 않음

## 나만의 테이블 생성하기
```sql
CREATE TABLE tablename
    (
        column_name data_type,
        column_name data_type
    );
    
/* 실제 사례 */
CREATE TABLE cats (
    name VARCHAR(100),
    age INT
);
    
CREATE TABLE dogs (
    name VARCHAR(50),
    breed VARCHAR(50),
    age INT
);
```

## 잘 동작하는지 확인

```sql
SHOW TABLES;

SHOW COLUMNS FROM <tablename>;

DESC <tablename>;

```



##  테이블 제거하기
```sql
DROP TABLE <tablename>;
DROP TABLE cats;

SHOW COLUMNS FROM cats;
DESC cats
```
